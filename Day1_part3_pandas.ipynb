### Basic operations 

```python
# A vector of numbers is represented in pandas by Series.

import pandas as pd
import numpy as np

s = pd.Series([1,3,5,np.nan,6,8])
s
```

```python
# Arrays are represented in pandas by DataFrames.

a = np.random.randn(6,4)
b = np.arange(6) + 1
c = ['A', 'B', 'C', 'D']

df = pd.DataFrame(a, index=b, columns=c)
df
```

```python
# We can also create DataFrame from a dictionary.

df2 = pd.DataFrame({
    'A' : 1.,
    'B' : pd.Timestamp('20130102'),
    'C' : pd.Series(1, index=list(range(4)), dtype='float32'),
    'D' : np.array([3] * 4,dtype='int32'),
    'E' : pd.Categorical(["test","train","test","train"]),
    'F' : 'foo'})

df2
```

```python
# We can also easily import data from a csv file.

df3 = pd.read_csv('data/iris.csv')
df3
```

```python
# In order to see only a few first rows, use head method.
# By default, 5 rows will be displayed.

df3.head(3)
```

```python
# To see the last rows, use tail method.
# By default, last 5 rows will be displayed.

df3.tail(4)
```

```python
# To quickly explore the data, we can use describe method.

df3.describe()
```

```python
# To find the flowers with the biggest sepal_length,
# we can sort our data and display first few items.

df3.sort_values(by='sepal_length', ascending=False).head()
```

```python
# We can select one column and rows from 5 to 9.

df3['sepal_length'][5:10]
```

```python
# We can select specific row.

df3[2:4]  # the same as df3.iloc[2:4]
```

```python
# We can change the index to something else.

df4 = df3[:4].copy()
df4.index = ['A', 'B', 'C', 'D']
df4
```

```python
# Now we can access the rows using either the row number or the row index.

df4.loc['C']  # the same as `df4[2]`.
```

```python
# We can easily select rows, for which sepal_length is larger than certain treshold.

df4[df4.sepal_length > 4.8]
```

```python
# We can add a new column.

df4['new'] = ['one',  np.NAN, np.NAN, 'four']
df4
```

```python
# We can clean our data, dropping rows with not-a-number values.

df4.dropna(how='any')
```

```python
# Or we can fill up the missing data.

df4.fillna(value='something')
```

```python
# We can create matrix, that will tell us where we have missing values.

pd.isna(df4)
```

```python
# To calculate mean for each column, use mean method.

df4.mean()
```

```python
# To calculate mean for each row, we can use mean method and specify which axis should be used.

df4.mean(1)
```

```python
# We can easily create a histogram of a selected column.

df4['petal_length'].value_counts()
```

```python
# We can group the data by values in one column.

df4.groupby('petal_length').mean()
```

```python
# We can plot it very easily.
# remember to use %matplotlib inline (once per sesion) before you plot anything!
%matplotlib inline

df3['petal_length'].value_counts().hist()  # note, we used df3 (whole dataset) not the subpart df4.
```

```python
# We can quickly create histograms for each column.

df3.hist()
```

```python
# We can plot density functions.

df3.plot(kind='density')
```

```python
# Or box plot.

df3.plot(kind='box')
```

```python
# We can calculate correlations in our datase.

correlations = df3.corr()
correlations
```

```python
# And we can visualize it.
# For basic information about matplotlib see https://matplotlib.org/users/pyplot_tutorial.html

import matplotlib.pyplot as plt

names = df3.columns
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,4,1)

ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)

ax.plot()
```

13. thal: 3 = normal; 6 = fixed defect; 7 = reversable defect
